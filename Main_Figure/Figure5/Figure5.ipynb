```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
feature = pd.read_csv('feature.txt', sep='\t', header=None, index_col=0)
labels_plate = pd.read_csv('labels_plates.txt', sep='\t', header=None)[1]
```

## Figure 5a

```python
from scale.plot import plot_embedding

legend_params = {'loc':'right', 'ncol':1, 'bbox_to_anchor':(1.0, 0.2), 'fontsize':8, 'markerscale':1.5}
```

### Full

```python
emb = plot_embedding(feature, labels_plate, legend_params=legend_params, return_emb=True)
np.savetxt('full_tsne.txt', emb)
```

### Plate-related

```python
index = [1, 2, 3, 4, 7, 9]
emb = plot_embedding(feature.iloc[:, index], labels_plate, legend_params=legend_params, return_emb=True)
np.savetxt('plate-related_tsne.txt', emb)
```

### Plate-independent

```python
index = [0, 5, 6, 8]
emb = plot_embedding(feature.iloc[:, index], labels_plate, legend_params=legend_params, return_emb=True)
np.savetxt('plate_independent_tsne.txt', emb)
```
